```python
def fact(n):
  if n==0:
    return 1
  return n*fact(n-1)
```

```python
def gamma(n):
  return fact(n-1)
```

```python
def beta(m,n):
  return(gamma(m)*gamma(n))/gamma(m+n)
```

```python
beta(2,5)
```
